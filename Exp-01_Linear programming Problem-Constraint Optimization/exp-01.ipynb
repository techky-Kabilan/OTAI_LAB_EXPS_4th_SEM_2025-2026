```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import linprog
# Objective function coefficients (Minimize Z = 50x + 100y)
c = [50, 100] # No need to negate since we are minimizing
# Coefficients for inequality constraints (converted to <= form)
A = [[-7, -2],
[-2, -12]]
# Right-hand side values for inequality constraints
b = [-28, -24] # Converted accordingly
# Bounds for variables (x, y ≥ 0)
x_bounds = (0, None)
y_bounds = (0, None)
# Solve the linear programming problem
res = linprog(c, A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds], method='highs')
# Extract the optimal solution
x_opt, y_opt = res.x
z_opt = res.fun # Since we are minimizing, no need to negate
print(f"Optimal solution: x = {x_opt}, y = {y_opt}, Z = {z_opt}")
# Plotting the constraints
x = np.linspace(0, 10, 400)
y1 = (28 - 7*x) / 2
y2 = (24 - 2*x) /12
plt.figure(figsize=(8, 8))
# Plot each constraint
plt.plot(x, y1, label=r'$7x + 2y \geq 28$')
plt.plot(x, y2, label=r'$2x + 12y \geq 24$', linestyle='--')
# Fill feasible region
y_feasible = np.maximum(y1, y2) # Take the maximum to satisfy both inequalities
plt.fill_between(x, y_feasible, 10, color='grey', alpha=0.5)
# Highlight the optimal solution point
plt.plot(x_opt, y_opt, 'ro', label=f'Optimal Solution (x={x_opt:.2f}, y={y_opt:.2f})')
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphical Solution of Linear Programming Problem (Minimization)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.optimize import linprog
# Define the coefficients of the objective function
# For example, maximize: z = 3x1 + 2x2
c = [-3, -2]

# Define the coefficients of the inequality constraints
# For example:
# 2x1 + x2 <= 20
# 4x1 - 5x2 <= 10
# x1, x2 >= 0 (non-negativity constraints)
A = [
[2, 1],
[4, -5]
]
# Define the right-hand side of the inequality constraints
b = [20, 10]
# Define bounds for variables (non-negativity constraints)
x_bounds = (0, None)
y_bounds = (0, None)
# Solve the linear programming problem
result = linprog(c, A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds], method='highs')
# Print the result
print('Optimal value:', -result.fun)
print('Values of x:', result.x)
```
